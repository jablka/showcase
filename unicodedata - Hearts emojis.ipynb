```python
# Using the unicodedata module, specifically its name function, 
# filter characters to show only those having 'HEART' in their name. 

# challenge from https://www.hackinscience.org/exercises/hearts-emojis
```

```python
import unicodedata
for i in range(1,230000):
    name = unicodedata.name(chr(i), 'pass')
    if 'HEART' in name:
        print(chr(i), end='')
```
